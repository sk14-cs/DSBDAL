```python
Aim:
Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for 
distribution of age with respect to each gender along with the information about whether they 
survived or not. (Column names : 'sex' and 'age')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='sex', y='age', hue='survived', data=titanic)
plt.title('Age Distribution with Respect to Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')

# Display the plot
plt.show()
```
